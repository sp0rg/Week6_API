## Results

1) Cities in the range from -20 to about 25 degrees latitude have the highest temperatures. Cities below or above that range drop off the further they away from 0 that they get. This isn't rocket surgery, folks.

2) Humidity and temperature are all over the place. There does not appear to be a strong relationship.

3) Similar to humidity, the clouds are all over the place.

4) Wind speed compared to latitude was pretty weak as well. Although there were more outliers of wind speed > 20mph from 20 degrees latitude and up.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import api_key


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File in csv format
# made an Output folder because the main folder got a bit messy
output_data_file = "Output/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

```python
#Defining URL and Units
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"

# Build partial query URL
#tried using an f-string and it didn't work at all for me - NOT SURE WHY
#seemed to work out using an alternate method
#query_url = f"{url}appid={api_key}&units={units}&q="

query_url = url + "appid=" + api_key + "&units=" + units + "&q="
#query_url
```

```python
# set up lists to hold reponse info
city_list=[] 
latitude=[]
longitude=[]
temperature=[]
humidity=[]
cloudiness=[]
wind_speed=[]

# City counter
total_cities_count=0

print("Beginning City Data Pull")
print("------------------------")

# Loop through the list of cities and perform a request for data on each
for city in cities:
    response = requests.get(query_url + city).json()
    try:
        #Used print(response)during trouble shooting - cities weren't pulling back at all
        #ended up being an issue with my url
        #print(response) 
        city_list.append(response['name'])
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        temperature.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        total_cities_count+=1
        time.sleep(1.1)
        #Added counter to print to follow along and get an idea of how
        #many iterations it will take to complete. Might have enough time
        #to bake brownies or rebuild a carburetor.
        print(f'Huzzah! City "{city}" has been found! This is found city {total_cities_count} of {len(cities)}.')
        
    except:
        print(f'City "{city}" not found. Move along, nothing to see here...')
    
print(f"Found and processed {total_cities_count} of {len(cities)} cities")
```

```python
# Make DataFrame for csv and plots
df=pd.DataFrame({'city':city_list,
               'latitude':latitude,
               'longitude':longitude,
                'temperature':temperature,
                'humidity':humidity,
                'cloudiness':cloudiness,
                'wind_speed':wind_speed})
```
